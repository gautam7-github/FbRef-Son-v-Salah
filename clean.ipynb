# SON

```python
import pandas as pd
```

```python
sonFP = pd.read_csv("./data/son_full_report.csv").drop("Unnamed: 0",axis=1)
```

```python
sonFP = sonFP.drop_duplicates()
```

```python
sonFP['Statistic'].values
```

```python
sonOverall = sonFP.loc[:10]
sonOverall.to_csv("./son/son_overall.csv",index=False)
sonOverall
```

```python
sonShooting = sonFP.loc[12:27]
sonShooting.to_csv("./son/son_shooting.csv",index=False,header=False)
sonShooting
```

```python
sonPassing = sonFP.loc[29:50]
sonPassing.to_csv("./son/son_passing.csv",index=False)
sonPassing
```

```python
sonOverall.set_index("Statistic").T.to_csv("./compare_data/son_overall.csv")
sonPassing.set_index("Statistic").T.to_csv("./compare_data/son_passing.csv")
sonShooting.set_index("Statistic").T.to_csv("./compare_data/son_shooting.csv")
```
